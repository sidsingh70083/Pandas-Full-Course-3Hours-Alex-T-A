### 1. Remove Duplicates

```python
df = df.drop_duplicates()
df
```

### 2. Dropping unwanted Columns

```python
df = df.drop(columns='Not_Useful_Column')
df
```

### 3. Cleaning Last_Name

```python
# df["Last_Name"]=df["Last_Name"].str.lstrip("/")
# df["Last_Name"]=df["Last_Name"].str.lstrip(".")
# df["Last_Name"]=df["Last_Name"].str.rstrip("_")

#this was one way of doing it; another and better way:

df["Last_Name"]=df["Last_Name"].str.strip("123./_")
df
```

### 4. Working with Phone Numbers

- First remove all the formats to make everything same; then change all the number format to same at once

```python
df["Phone_Number"]=df["Phone_Number"].str.replace('[^a-zA-Z0-9]','',regex=True)
```

- now we will use a lambda function for applying the same format to each phone number

- [Note]: you have to convert the numbers into str to format then as you know '-' and any other symbol, concatenation with number is not allowed

if you use str(df["Phone_Number"]), it will convert the complete column at once in to str, but we want to change each number/value (each row)

```python
str(df["Phone_Number"])
```

```python
df["Phone_Number"]=df["Phone_Number"].apply(lambda x:str(x))
```

```python
df["Phone_Number"]=df["Phone_Number"].apply(lambda x: x[0:3]+"-"+x[3:6]+"-"+x[6:10])
df
```

```python
df["Phone_Number"]=df["Phone_Number"].str.replace('nan--',"")
df["Phone_Number"]=df["Phone_Number"].str.replace('Na--',"")
df
```

```python
df[['Street_Address','State','Zip_Code']]=df["Address"].str.split(',',n=2,expand=True)
df
```

### 5. Paying/Do_Not_Contact

```python
df["Paying Customer"]=df["Paying Customer"].str.replace('Yes','Y',regex=True)
df["Paying Customer"]=df["Paying Customer"].str.replace('No','N',regex=True)
df
```

```python
df["Do_Not_Contact"]=df["Do_Not_Contact"].str.replace('Yes','Y',regex=True)
df["Do_Not_Contact"]=df["Do_Not_Contact"].str.replace('No','N',regex=True)
df
```

- Cleaning up N/a and NaN

```python
df=df.fillna('')
df
```

- Removing options with Do not contact as: Yes or with no phone number

```python
for x in df.index:
    if df.loc[x,"Do_Not_Contact"]=='Y':
        df.drop(x,inplace=True)

df
```

```python
for x in df.index:
    if df.loc[x,"Phone_Number"]=='':
        df.drop(x,inplace=True)

df

# Method 2 of doing this one (dropping null values) was using 
# df.dropna(subset="Phone_Number",inplace=True)
```

```python
df.drop('Address',axis=1,inplace=True)
df
```
