# Merge

```python
df1.merge(df2,how='inner') # it automatically takes few things into consideration; check the below code, it also gives the same results.
```

```python
df1.merge(df2, how='inner',on=['FellowshipID','FirstName'])
```

```python
df1.merge(df2, how='outer') # again the same thing is true for this one too

#  df1.merge(df2, how='outer',on=['FellowshipID','FirstName'])
```

```python
df1.merge(df2, how='left') #everything in left and overlapping
```

```python
df1.merge(df2, how='right') #everything in right and overlapping
```

### Cross Merge

* It's like a nested loop, for each value in left df, it give/compares for all value in df2

```python
df1.merge(df2, how='cross')
```

***

# Joins

```python
df1.join(df2,on="FellowshipID",how='outer',lsuffix='_Left',rsuffix='_Right')
```

* Joins usually works well with index. ->

```python
df3 = df1.set_index('FellowshipID').join(df2.set_index('FellowshipID'), how='outer',lsuffix='_Left',rsuffix='_Right')
df3
```

# Concatenate

* Difference between Join and Conc. is that: It's like putting one df on top of the other, rather than putting next to one another.     

```python
pd.concat([df1,df2],join='outer', axis=1) 
# by axis it means that now conc. will take effect on columns; as you can see columns are now repeating/stacking after 1 df is finished
# axis={1:'Columns/x-axis',0:'Rows/y-axis'}
```
