```python
df.plot(kind='line',subplots=True, xlabel="daily records",ylabel="scores")
```

```python
df.plot.barh(stacked=True)
```

```python
df.plot.scatter(x='Flavor Rating',y='Overall Rating', s=50,c='Red')
```

```python
df.plot.hist(bins=10)
```

```python
df.boxplot() 
# lower line: min value
# bottom line of box: 25%
# middle line of box: 50%
# top line of box: 76%
# top line: max value
```

* you can use figsize=(,) for the size of the complete chart

```python
df.plot.area(figsize=(10,3))
```

```python
df.plot.pie(y='Flavor Rating', figsize=(3,3))
```
